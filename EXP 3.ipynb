```python

import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload image
uploaded = files.upload()
file = list(uploaded.keys())[0]

# Step 2: Read image
img = cv2.imread(file)

# Step 3: Apply CLEARLY VISIBLE blur
blur = cv2.GaussianBlur(img, (31,31), 10)

# Step 4: Show Original & Blurred
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Blurred Image")
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

# Step 5: Histogram Graph
gray = cv2.cvtColor(blur, cv2.COLOR_BGR2GRAY)

plt.figure()
plt.title("Histogram of Blurred Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.hist(gray.ravel(), bins=256)
plt.show()

# Step 6: Difference (PROOF of blur)
diff = cv2.absdiff(img, blur)

plt.figure()
plt.title("Difference Image (Proof of Blur)")
plt.imshow(cv2.cvtColor(diff, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
